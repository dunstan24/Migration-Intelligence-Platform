```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

def create_sequences(data, target, time_steps=10):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i : (i + time_steps)])
        y.append(target[i + time_steps])
    return np.array(X), np.array(y)

def undersample_sequences(X, y):
    # Cari indeks untuk masing-masing kelas
    indices_class_0 = np.where(y == 0)[0]
    indices_class_1 = np.where(y == 1)[0]  # Misal Invited
    # ... ulangi untuk kelas lain

    # Tentukan jumlah sampel minimum (misal disamakan dengan jumlah kelas tersedikit)
    n_samples = len(indices_class_1)

    # Ambil sampel acak dari kelas mayoritas
    random_indices_0 = np.random.choice(indices_class_0, n_samples)

    # Gabungkan kembali
    new_indices = np.concatenate([random_indices_0, indices_class_1])

    return X[new_indices], y[new_indices]

# Asumsi df_encoded adalah DataFrame hasil Binary Encoding Anda
# Fitur = 9 kolom biner + fitur numerik lainnya


```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_3d, y_3d, test_size=0.2, shuffle=False
)
```

```python
model = Sequential(
    [
        # Layer LSTM pertama
        LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
        Dropout(0.2),
        # Layer LSTM kedua
        LSTM(32, return_sequences=False),
        Dropout(0.2),
        # Fully Connected Layer
        Dense(16, activation="relu"),
        # Output Layer (Gunakan 'sigmoid' untuk binary, 'softmax' untuk multi-class)
        Dense(1, activation="sigmoid"),
    ]
)

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
    
)
model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2,  # Mengambil 20% data untuk validasi
    verbose=1,
    callbacks=early_stop
)
```
